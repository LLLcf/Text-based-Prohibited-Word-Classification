```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# 设置中文字体，确保中文正常显示
plt.rcParams["font.family"] = ["SimHei", "WenQuanYi Micro Hei", "Heiti TC"]
plt.rcParams['axes.unicode_minus'] = False  # 解决负号显示问题

# 假设data_train是你的训练集DataFrame，包含"文本"列
# 计算所有文本的长度
train_text_lengths = data['文本'].apply(len)

# 计算基本统计信息
train_max_len = train_text_lengths.max()
train_min_len = train_text_lengths.min()
train_mean_len = train_text_lengths.mean()
train_median_len = train_text_lengths.median()
train_90p_len = train_text_lengths.quantile(0.9)

print(f'训练集文本长度统计:')
print(f' - 最大值: {train_max_len}')
print(f' - 最小值: {train_min_len}')
print(f' - 平均值: {train_mean_len:.2f}')
print(f' - 中位数: {train_median_len}')
print(f' - 90%分位数: {train_90p_len}')

# 创建画布
plt.figure(figsize=(12, 6))

# 计算密度曲线数据
density = stats.gaussian_kde(train_text_lengths)
x_range = np.linspace(train_text_lengths.min(), train_text_lengths.max(), 200)
density_values = density(x_range)

# 绘制密度曲线
plt.plot(x_range, density_values, color='#FF5722', linewidth=3, label='训练集文本长度密度分布')

# 填充曲线下方区域，增强视觉效果
plt.fill_between(x_range, density_values, alpha=0.3, color='#FF5722')

# 添加统计参考线
plt.axvline(train_mean_len, color='r', linestyle='dashed', linewidth=2, 
            label=f'平均值: {train_mean_len:.2f}')
plt.axvline(train_median_len, color='g', linestyle='dashed', linewidth=2, 
            label=f'中位数: {train_median_len}')

# 设置标题和标签
plt.title('训练集文本长度密度分布', fontsize=16)
plt.xlabel('文本长度', fontsize=14)
plt.ylabel('密度', fontsize=14)

# 添加图例和网格
plt.legend(fontsize=12)
plt.grid(axis='x', alpha=0.3)

# 调整布局
plt.tight_layout()

# 显示图像
plt.show()
```
